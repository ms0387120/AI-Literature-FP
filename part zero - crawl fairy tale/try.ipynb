### 先試爬「小紅帽」

```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.grimmstories.com/zh/grimm_tonghua/xiaohongmao'
resp = requests.get(url)
resp.encoding = 'utf-8'
soup = BeautifulSoup(resp.text, 'html.parser')

corpus = soup.find('div', attrs={'class':'text'})
print(corpus.text)
```

### 簡體轉繁體

```python
from langconv import *
from langconv import *

def Simplified2Traditional(sentence):
    
    sentence = Converter('zh-hant').convert(sentence)
    return sentence
```

```python
print(Simplified2Traditional(corpus.text))
```

### 結果寫入檔案

```python
z = ['小紅帽']

# 開啟檔案
filename = str(z[0])
f = open(filename+".txt", "a")
 
# 寫入 This is a testing! 到檔案
f.write(Simplified2Traditional(corpus.text))
 
# 關閉檔案
f.close()
```
